# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urunbilgisikaydet():
    urunadı=input("ÜRÜN ADI:")
    urunfiyat=input("ÜRÜN FİYATI:")
    with open("urunler.txt","a") as file:
        file.write(f"Ad:{urunadı},\n Fiyat:{urunfiyat}")
        print("Ürün bilgileri kaydedildi.")
urunbilgisikaydet()
print(urunbilgisikaydet)
````
