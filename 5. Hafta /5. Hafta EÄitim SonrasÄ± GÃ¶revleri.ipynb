# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python
def hesapla_son_hiz(vi,a,t):
    vf=(vi+a)*t
    return vf
ilkhiz=15
ivme=3
zaman=6
son_hiz=hesapla_son_hiz(ilkhiz,ivme,zaman)
print(f"son hız:{son_hiz} m/s")
````

# Problem 2

1'den 1000'e kadar olan sayılardan mükemmel sayı olanları ekrana yazdırın. Bunun için bir sayının mükemmel olup olmadığını dönen bir tane fonksiyon yazın.

Bir sayının bölenlerinin toplamı kendine eşitse bu sayı mükemmel bir sayıdır. Örnek olarak 6 mükemmel bir sayıdır (1 + 2 + 3 = 6).

````python
def tambolen(n):
    bolenler=[]
    for i in range(1,n):
        if n % i == 0:
            bolenler.append(i)
    return bolenler
def mukemmelsayimi(n):
    return sum(tambolen(n))==n
for i in range(1,1001):
    if mukemmelsayimi(i):
     print(f"{i} mükemmel bir sayıdır")
````

# Problem 3

````python
def pisagor():
    for a in range(1,101):
        for b in range(a,101):
            hipo=a**2+b**2
            c=int(hipo**0.5)
            if  c**2==hipo and c<=100:
                print(f"{a},{b},{c} bir pisagor ücgenidir.")
pisagor()
````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python
duvaryuksekligi=float(input("Duvar yüksekliğini giriniz:"))
duvargenisligi=float(input("Duvar genişliğini giriniz:"))
kbkaplama=5
def kutusayısıhesapla(duvaryuksekligi,duvargenisligi):
    alan=duvaryuksekligi*duvargenisligi
    kutusayısı=alan/kbkaplama
    kutusayısı=int(kutusayısı) if kutusayısı==int(kutusayısı) else int(kutusayısı)+1
    return kutusayısı  
kutusayısı= kutusayısıhesapla(duvargenisligi,duvaryuksekligi)  
print(f"Kutu Sayısı:{kutusayısı}")
````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
kelime=str(input("Bir kelime giriniz"))
def harfsayisi(kelime):
    unluharfler="aeiıoöuü"
    unlusayısı=0
    unsuzsayısı=0
    for harf in kelime:
        if harf.lower() in unluharfler:
            unlusayısı+=1
        else:
            unsuzsayısı+=1
    return unlusayısı,unsuzsayısı
unlusayısı,unsuzsayısı=harfsayisi(kelime)
print(f"ünlü harf sayısı:{unlusayısı}")
print(f"ünsüz harf sayısı:{unsuzsayısı}")

````

# Problem 6

Verilen bir listeden çift sayıları yazdıran bir fonksiyon yazınız.


````python
sayilarlistesi=input("bir sayı listesi giriniz")
def sayilar(sayilar):
    ciftsayilar="02468"
    teksayilar="13579"
    ciftrakamsayisi=0
    tekrakamsayisi=0
    for sayi in sayilar:
        if sayi in ciftsayilar:
            ciftrakamsayisi += 1
        else:
            tekrakamsayisi += 1
    return ciftrakamsayisi,tekrakamsayisi
ciftrakamsayisi,tekrakamsayisi = sayilar(sayilarlistesi)
print(f"çift rakam sayısı:{ciftrakamsayisi}")
print(f"tek rakam sayısı:{tekrakamsayisi}")
````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
def toplam(p):
    if p==0:
        return 0
    else:
        return p+ toplam(p-1)
sonuç=toplam(10)
print(f"0'dan 10'a kadar olan sayıların toplamı:{sonuç}")
````
